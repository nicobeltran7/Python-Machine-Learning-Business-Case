```python
from sklearn.linear_model import LinearRegression
Regression_testing = LinearRegression()
```

```python
target = ["SalePrice"]
predictors = ["Total Area","Year Built"]
```

```python
Regression_testing.fit(train[predictors], train[target])
```

```python
#Train/test strategy & baseline model

#Time-based split: train on Yr Sold < 2009, test on â‰¥ 2009 to mimic future prediction and prevent leakage from later years.
#Model: LinearRegression with two features (Total Area, Year Built) as a baseline.
#Goal: establish a transparent starting point before adding regularization, interactions, or log-transform.
```

```python
target_predictions = Regression_testing.predict(test[predictors]).round(2)
target_predictions[:20]
```
