```python
df["Yr Sold"].unique()
```

```python
train = df[df["Yr Sold"]<2009].copy()
test = df[df["Yr Sold"]>=2009].copy()
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn.linear_model import LinearRegression
Regression_testing = LinearRegression()
```

```python
target = ["SalePrice"]
predictors = ["Total Area","Year Built"]
```

```python
Regression_testing.fit(train[predictors], train[target])
```

```python
target_predictions = Regression_testing.predict(test[predictors]).round(2)
target_predictions
```

```python
test["target_predictions"] = target_predictions
test
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
errors = mean_absolute_error(test["SalePrice"],test["target_predictions"])
errors
```

```python
df.describe()["SalePrice"]
```

```python
difference = (test["SalePrice"]-test["target_predictions"]).abs()
difference
```

```python
difference_by_neighborhood = difference.groupby(test["Neighborhood"]).mean()
difference_by_neighborhood
```

```python
saleprice_by_neighborhood = test["SalePrice"].groupby(test["Neighborhood"]).mean()
```

```python
error_difference = difference_by_neighborhood/saleprice_by_neighborhood
error_difference.sort_values()
```

```python
error_difference.plot.hist()
plt.show()
```
